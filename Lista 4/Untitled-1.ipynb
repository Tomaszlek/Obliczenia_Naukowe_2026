```python
#importy standardowe
import sympy as sp
from sympy import symbols, diff, solve, sqrt, Matrix, simplify, lambdify
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sympy.plotting import plot_implicit, plot3d
import math

# Inicjalizacja pretty print dla SymPy
sp.init_printing(use_unicode=True)

#definicja symboli
x, y, z, l = symbols('x y z l')
print('Importy prawidłowe - gotowe do zadania')
```

## Zadanie 1 - Ekstremum warunkowe

**Treść zadania:**
Wzgórze ma kształt stożka, który w układzie XYZ ma równanie: $z = 6 - \sqrt{3x^2 + 3y^2}$

Turysta porusza się po zboczu wzgórza w ten sposób, że rzut jego trasy na płaszczyznę XY jest prostą o równaniu: $y = x + 1$

**Szukane:** Współrzędne najwyżej położonego punktu trasy turysty.

Będziemy rozwiązywać to zadanie trzema metodami:

```python
# ============================================================================
# I SPOSÓB - METODA MNOŻNIKÓW LAGRANGE'A
# ============================================================================

# Definicja funkcji celu i warunku
f = 6 - sqrt(3*x**2 + 3*y**2)  # funkcja wysokości wzgórza
g = y - x - 1  # warunek: y = x + 1, czyli g(x,y) = y - x - 1 = 0

print("=" * 70)
print("I SPOSÓB - METODA MNOŻNIKÓW LAGRANGE'A")
print("=" * 70)
print("\nFunkcja celu:")
print(f"f(x,y) = {f}")
print("\nWarunek:")
print(f"g(x,y) = {g} = 0")

# Budujemy funkcję Lagrange'a: L(x,y,λ) = f(x,y) - λ*g(x,y)
L = f - l*g
print("\nFunkcja Lagrange'a:")
print(f"L(x,y,λ) = {L}")

# Wyznaczamy pochodne pierwszego rzędu
dLdl = diff(L, l)
dLdx = diff(L, x)
dLdy = diff(L, y)

print("\nPochodne pierwszego rzędu:")
print(f"∂L/∂λ = {dLdl}")
print(f"∂L/∂x = {dLdx}")
print(f"∂L/∂y = {dLdy}")

# Rozwiązujemy układ równań: ∇L = 0
print("\nRozwiązywanie układu równań...")
critical_points = solve([dLdl, dLdx, dLdy], [x, y, l])
print(f"\nPunkty krytyczne: {critical_points}")

# Proces liczenia dla każdego punktu krytycznego
for i, cp in enumerate(critical_points):
    print(f"\n{'─' * 70}")
    print(f"Punkt krytyczny #{i+1}:")
    x0, y0, l0 = cp[x], cp[y], cp[l]
    print(f"x₀ = {x0}, y₀ = {y0}, λ₀ = {l0}")
    
    # Wyznaczamy macierz Hessego dla funkcji Lagrange'a
    # H = [0     g_x   g_y  ]
    #     [g_x   L_xx  L_xy ]
    #     [g_y   L_yx  L_yy ]
    
    dg_dx = diff(g, x)
    dg_dy = diff(g, y)
    dL_dxx = diff(L, x, 2)
    dL_dyy = diff(L, y, 2)
    dL_dxy = diff(L, x, y)
    
    # Macierz brzegowa Hessego
    H = Matrix([
        [0, dg_dx, dg_dy],
        [dg_dx, dL_dxx, dL_dxy],
        [dg_dy, dL_dxy, dL_dyy]
    ])
    
    print(f"\nMacierz brzegowa Hessego Δ₂:")
    H_vals = H.subs({x: x0, y: y0})
    print(H_vals)
    
    # Wyznacznik Hessego
    delta2 = H_vals.det()
    print(f"\ndet(Δ₂) = {delta2}")
    
    # Ocena ekstremum
    if delta2 > 0:
        print("→ MAKSIMUM warunkowe")
    elif delta2 < 0:
        print("→ MINIMUM warunkowe")
    else:
        print("→ Brak ekstremum (punkt osobliwy)")
    
    # Wartość funkcji w punkcie
    f_val = f.subs({x: x0, y: y0})
    print(f"f({x0}, {y0}) = {f_val} ≈ {float(f_val.evalf()):.6f}")

print("\n" + "=" * 70)
```

```python
# ============================================================================
# II SPOSÓB - METODA GRAFICZNA
# ============================================================================

print("\n" + "=" * 70)
print("II SPOSÓB - METODA GRAFICZNA")
print("=" * 70)

# Warstwice (kontury) funkcji f(x,y) = 6 - √(3x² + 3y²)
# Warstwica to zbiór punktów, gdzie f(x,y) = const = c
# 6 - √(3x² + 3y²) = c
# √(3x² + 3y²) = 6 - c
# 3x² + 3y² = (6-c)²
# x² + y² = ((6-c)/√3)²

print("\nRównanie warstwic funkcji f:")
print("Warstwice to elipsy postaci: x² + y² = r²")
print("gdzie r = (6-c)/√3, c - stała na warstwicy")

# Rysowanie warstwic i warunku
fig, ax = plt.subplots(figsize=(12, 10))

# Siatka punktów dla płaszczyzny XY
x_vals = np.linspace(-5, 5, 500)
y_vals = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x_vals, y_vals)

# Funkcja wysokości
Z = 6 - np.sqrt(3*X**2 + 3*Y**2)

# Warstwice funkcji f
levels = [1, 2, 3, 4, 4.5, 5, 5.5, 5.8]
contours = ax.contour(X, Y, Z, levels=levels, colors='blue', alpha=0.6, linewidths=1.5)
ax.clabel(contours, inline=True, fontsize=8)

# Prosta warunku: y = x + 1
x_line = np.linspace(-5, 4, 100)
y_line = x_line + 1
ax.plot(x_line, y_line, 'r-', linewidth=2.5, label='Warunek: y = x + 1')

# Znalezienie punktu gdzie warstwica ma jeden punkt wspólny z prostą
# Zamieniamy y = x + 1 do równania warstwicy
# x² + (x+1)² = r² = ((6-c)/√3)²
# x² + x² + 2x + 1 = ((6-c)/√3)²
# 2x² + 2x + 1 = ((6-c)/√3)²

# Dla ekstremum, dyskryminanta tego równania powinna być = 0
# Podstawiamy y = x + 1 do funkcji f
a_coeff = 3
b_coeff = 3

# f(x, x+1) = 6 - √(3x² + 3(x+1)²)
#           = 6 - √(3x² + 3x² + 6x + 3)
#           = 6 - √(6x² + 6x + 3)

expr = 6*x**2 + 6*x + 3

print("\nPodstawienie y = x + 1 do wyrażenia pod pierwiastkiem:")
print(f"3x² + 3(x+1)² = 3x² + 3(x² + 2x + 1) = 6x² + 6x + 3")

# Szukamy ekstremum f(x, x+1)
f_reduced = 6 - sqrt(expr)
print(f"\nFunkcja zredukowana: f(x, x+1) = {f_reduced}")

# Pochodna
df_reduced = diff(f_reduced, x)
print(f"Pochodna: df/dx = {df_reduced}")

# Punkty krytyczne
critical_x = solve(df_reduced, x)
print(f"\nPunkty krytyczne: {critical_x}")

# Dla każdego punktu krytycznego
for x_crit in critical_x:
    if x_crit.is_real:
        y_crit = x_crit + 1
        f_crit = f_reduced.subs(x, x_crit)
        print(f"\nPunkt: ({x_crit}, {y_crit})")
        print(f"Wartość funkcji: {f_crit} ≈ {float(f_crit.evalf()):.6f}")
        
        # Zaznaczenie na wykresie
        ax.plot(float(x_crit.evalf()), float(y_crit.evalf()), 'go', markersize=12, 
                label=f'Ekstremum: ({float(x_crit.evalf()):.3f}, {float(y_crit.evalf()):.3f})')
        
        # Rysowanie warstwicy przez ten punkt
        c_val = float(f_crit.evalf())
        levels_special = [c_val]
        contours_special = ax.contour(X, Y, Z, levels=levels_special, colors='green', 
                                     linewidths=3, linestyles='--')

ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_title('Metoda graficzna - Warstwice funkcji i warunek', fontsize=14, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10, loc='best')
ax.set_xlim(-5, 4)
ax.set_ylim(-5, 5)
ax.set_aspect('equal')

plt.tight_layout()
plt.show()

print("\n" + "=" * 70)
```

```python
# ============================================================================
# III SPOSÓB - SPROWADZENIE DO EKSTREMUM LOKALNEGO
# ============================================================================

print("\n" + "=" * 70)
print("III SPOSÓB - SPROWADZENIE DO EKSTREMUM LOKALNEGO")
print("=" * 70)

# Z warunku y = x + 1, eliminujemy zmienną y
# Podstawiamy do funkcji f

print("\nKrok 1: Eliminacja zmiennej y")
print("Z warunku: y = x + 1")
print("Podstawiamy do f(x,y) = 6 - √(3x² + 3y²)")

# Funkcja z wyrażona w zależności od x
f_of_x = 6 - sqrt(3*x**2 + 3*(x + 1)**2)
f_of_x_simplified = simplify(f_of_x)

print(f"\nf(x, x+1) = {f_of_x}")
print(f"Uproszczone: f(x, x+1) = {f_of_x_simplified}")

# Rozwinięcie wyrażenia
print("\nRozpisu:")
print(f"3x² + 3(x+1)² = 3x² + 3(x² + 2x + 1) = 3x² + 3x² + 6x + 3 = 6x² + 6x + 3")
print(f"f(x, x+1) = 6 - √(6x² + 6x + 3)")

# Teraz szukamy ekstremum lokalnego funkcji jednej zmiennej
f_single = 6 - sqrt(6*x**2 + 6*x + 3)

print("\nKrok 2: Wyznaczenie ekstremów funkcji jednej zmiennej")
print(f"f(x) = {f_single}")

# Pierwsza pochodna
df_dx = diff(f_single, x)
print(f"\nPierwsza pochodna:")
print(f"f'(x) = {df_dx}")

# Uproszczenie
df_dx_simplified = simplify(df_dx)
print(f"Uproszczone: f'(x) = {df_dx_simplified}")

# Punkty krytyczne
print("\nRozwiązywanie f'(x) = 0:")
critical_points_x = solve(df_dx, x)
print(f"Punkty krytyczne: {critical_points_x}")

# Druga pochodna
d2f_dx2 = diff(df_dx, x)
print(f"\nDruga pochodna:")
print(f"f''(x) = {d2f_dx2}")

# Analiza każdego punktu krytycznego
print("\n" + "─" * 70)
print("Krok 3: Analiza typu ekstremum")
print("─" * 70)

max_point = None
max_value = -float('inf')

for x_crit in critical_points_x:
    if x_crit.is_real:
        print(f"\nPunkt krytyczny: x₀ = {x_crit}")
        
        # Obliczenie wartości drugiej pochodnej
        d2f_val = d2f_dx2.subs(x, x_crit)
        print(f"f''({x_crit}) = {d2f_val} ≈ {float(d2f_val.evalf()):.6f}")
        
        # Określenie typu ekstremum
        if d2f_val < 0:
            print("→ MAKSIMUM lokalne")
            ekstremum_type = "MAKSIMUM"
        elif d2f_val > 0:
            print("→ MINIMUM lokalne")
            ekstremum_type = "MINIMUM"
        else:
            print("→ Punkt osobliwy (test nierozstrzygający)")
            ekstremum_type = "???"
        
        # Obliczenie odpowiadającego y
        y_crit = x_crit + 1
        print(f"y₀ = x₀ + 1 = {y_crit}")
        
        # Wartość funkcji f w punkcie
        z_crit = f_single.subs(x, x_crit)
        z_crit_val = float(z_crit.evalf())
        print(f"f({x_crit}, {y_crit}) = {z_crit}")
        print(f"Wartość liczbowa: {z_crit_val:.6f}")
        
        # Śledzenie maksimum
        if ekstremum_type == "MAKSIMUM":
            if z_crit_val > max_value:
                max_value = z_crit_val
                max_point = (x_crit, y_crit, z_crit)

print("\n" + "=" * 70)
print("PODSUMOWANIE ROZWIĄZANIA")
print("=" * 70)

if max_point:
    x0, y0, z0 = max_point
    print(f"\nNajwyżej położony punkt trasy turysty:")
    print(f"x = {x0} ≈ {float(x0.evalf()):.6f}")
    print(f"y = {y0} ≈ {float(y0.evalf()):.6f}")
    print(f"z = {z0} ≈ {float(z0.evalf()):.6f}")
    
    # Weryfikacja warunku
    y_check = y0 - x0 - 1
    print(f"\nWeryfikacja warunku y = x + 1:")
    print(f"y - x - 1 = {simplify(y_check)} ✓")

print("\n" + "=" * 70)
```

```python
# ============================================================================
# PORÓWNANIE WYNIKÓW Z TRZECH METOD
# ============================================================================

print("PORÓWNANIE WYNIKÓW Z TRZECH METOD")
print("=" * 70)
print("\nWszystkie trzy metody powinny dać ten sam wynik:\n")

# Wyznaczenie dokładnego wyniku
x_solution = critical_points_x[0]
y_solution = x_solution + 1
z_solution = 6 - sqrt(3*x_solution**2 + 3*y_solution**2)

print(f"Dokładne współrzędne punktu maksymalnego:")
print(f"  x = {x_solution}")
print(f"  y = {y_solution}")
print(f"  z = {z_solution}")

print(f"\nWartości przybliżone:")
print(f"  x ≈ {float(x_solution.evalf()):.8f}")
print(f"  y ≈ {float(y_solution.evalf()):.8f}")
print(f"  z ≈ {float(z_solution.evalf()):.8f}")

# Weryfikacja
print(f"\nWeryfikacja:")
print(f"  1. Czy punkt leży na prostej y = x + 1?")
print(f"     {float(y_solution.evalf()):.8f} ≈ {float(x_solution.evalf()):.8f} + 1")
print(f"     {float(y_solution.evalf()):.8f} = {float((x_solution + 1).evalf()):.8f} ✓")

print(f"\n  2. Czy punkt leży na wzgórzu z = 6 - √(3x² + 3y²)?")
z_check = 6 - np.sqrt(3*float(x_solution.evalf())**2 + 3*float(y_solution.evalf())**2)
print(f"     z = 6 - √(3·{float(x_solution.evalf()):.6f}² + 3·{float(y_solution.evalf()):.6f}²)")
print(f"     z ≈ {z_check:.8f}")
print(f"     {float(z_solution.evalf()):.8f} = {z_check:.8f} ✓")

print("\n" + "=" * 70)
print("ODPOWIEDŹ KOŃCOWA")
print("=" * 70)
print(f"\nNajwyżej położony punkt trasy turysty ma współrzędne:")
print(f"\n  ({x_solution}, {y_solution}, {z_solution})")
print(f"\nLub w przybliżeniu:")
print(f"  ({float(x_solution.evalf()):.6f}, {float(y_solution.evalf()):.6f}, {float(z_solution.evalf()):.6f})")

print("\n" + "=" * 70)
```
