```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
function = -5 * x * (2 * x - 3) ** (sp.Rational(1, 5))
```

```python
# funkcja pomocnicza do formatowania wyników do 8 miejsc po przecinku
def format_result(result):
    result = sp.N(result, 50)
    return f"{float(result):.8f}"
```

### zadanie 1
### Wyznacz dziedzinę funkcji i asymptoty, jeśli są.


```python
x = sp.symbols('x')
print('Dziedzina: R (wszystkie liczby rzeczywiste)')
lim_pos = sp.limit(function, x, sp.oo)
lim_neg = sp.limit(function, x, -sp.oo)
print('Granica przy +oo:', lim_pos)
print('Granica przy -oo:', lim_neg)
print('Wniosek: funkcja rośnie jak ~ C * x^(6/5), więc nie ma asymptoty poziomej ani ukośnej (wykładnik > 1).')
```

### zadanie 2 
### Narysuj wykres funkcji.


```python
def real_fifth_root(z):
    return np.sign(z) * np.abs(z) ** (1/5)

xs = np.linspace(-10, 10, 1000)
ys = -5 * xs * real_fifth_root(2 * xs - 3)

plt.figure(figsize=(8,5))
plt.plot(xs, ys, label='f(x) = -5 x (2x-3)^{1/5}')
plt.axhline(0, color='k', linewidth=0.6)
plt.axvline(0, color='k', linewidth=0.6)
plt.xlim(-10,10)
plt.ylim(np.nanmin(ys)*1.1, np.nanmax(ys)*1.1)
plt.grid(True)
plt.legend()
plt.title('Wykres funkcji')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
# Liczymy dlugosc łuku dla x należącego do przedziału [2, 3]
pochodna_funkcji = sp.diff(function, x)
dlugosc_luku = sp.integrate(sp.sqrt(1 + pochodna_funkcji**2), (x, 2, 3))

dlugosc_luku

print(f"Dlugosc łuku tej funkcji to: {format_result(dlugosc_luku)}")
```

### zadanie 10

```python
# Teraz objetosc bryly obrotowej
# Ona jest obrocona wokol osi OX, 0 <= y <= f(x), 2 <= x <= 3

objetosc_kreciolka = sp.integrate(sp.pi * function**2, (x, 2, 3))

objetosc_kreciolka
```
