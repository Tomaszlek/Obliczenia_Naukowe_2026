## Zadanie 1

```python
import sympy as sp

x, y = sp.symbols('x y', real=True)
```

### Zadanie 1

```python
f1 = 3*x**4 + 2*x*y - 3*x - 6*y**2 - 5
I1 = sp.simplify(sp.integrate(sp.integrate(f1, (y, 1, 7)), (x, -3, 4)))

f2 = x**2 * y * sp.exp(x*y)
I2 = sp.simplify(sp.integrate(sp.integrate(f2, (x, 0, 1)), (y, 1, 2)))

I3 = sp.simplify(sp.integrate(sp.integrate(x * sp.cos(x**2 + y), (y, 0, sp.pi)), (x, -sp.sqrt(sp.pi)/2, 0)))

I4 = sp.simplify(sp.integrate(sp.integrate(sp.exp(x) * sp.cos(2*x - 3*y), (y, x - 1, -x + 1)), (x, 0, 1)))

for name, val in [('I1', I1), ('I2', I2), ('I3', I3), ('I4', I4)]:
    print(name)
    display(val)
```

### Zadanie 2

```python
r, theta = sp.symbols('r theta', real=True)
f21 = (1 - r**2) / (1 + r**2) * r
I21 = sp.simplify(sp.integrate(sp.integrate(f21, (r, 1, 2)), (theta, -sp.pi/2, sp.pi/2)))

f22 = sp.log(1 + r**2) * r
I22 = sp.simplify(sp.integrate(sp.integrate(f22, (r, 0, 1)), (theta, -sp.pi/4, sp.pi/4)))

for name, val in [('I21', I21), ('I22', I22)]:
    print(name)
    display(val)
```

### Zadanie 3

```python
area_d = sp.pi
srednia = sp.simplify(sp.integrate(sp.integrate((5 - r**2) * r, (r, 0, 1)), (theta, 0, 2*sp.pi)) / area_d)
print('srednia')
display(srednia)
```

### Zadanie 4

```python
A4 = sp.simplify(sp.integrate(x, (x, 0, 1)) + sp.integrate(2/x - 1, (x, 1, 2)))
print('A4')
display(A4)
```

### Zadanie 5

```python
vol = sp.simplify(sp.integrate(sp.integrate(x**2 + y**2, (y, 0, 4 - x)), (x, 1, 4)))
print('vol')
display(vol)
```

### Zadanie 6

```python
f6 = 25 - 2*x - 3*y
A6 = sp.simplify(sp.integrate(sp.integrate(sp.sqrt(1 + sp.diff(f6, x)**2 + sp.diff(f6, y)**2), (y, x - 5, 0)), (x, 2, 4)))
print('A6')
display(A6)
```

```python
m = sp.simplify(sp.integrate(sp.integrate(1, (y, 3, 8)), (x, 1, 5)))
MX1 = sp.simplify(sp.integrate(sp.integrate(y, (y, 3, 8)), (x, 1, 5)))
MY1 = sp.simplify(sp.integrate(sp.integrate(x, (y, 3, 8)), (x, 1, 5)))
MX2 = sp.simplify(sp.integrate(sp.integrate(y**2, (y, 3, 8)), (x, 1, 5)))
MY2 = sp.simplify(sp.integrate(sp.integrate(x**2, (y, 3, 8)), (x, 1, 5)))
I0 = sp.simplify(sp.integrate(sp.integrate(x**2 + y**2, (y, 3, 8)), (x, 1, 5)))
xc = sp.simplify(MY1 / m)
yc = sp.simplify(MX1 / m)
for name, val in [('m', m), ('MX1', MX1), ('MY1', MY1), ('MX2', MX2), ('MY2', MY2), ('I0', I0), ('xc', xc), ('yc', yc)]:
    print(name)
    display(val)
```
