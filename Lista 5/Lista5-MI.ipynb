```python
import sympy as sp

x, y = sp.symbols('x y', real=True)
```

Zadanie 1 – całki podwójne

```python
expr1 = 3*x**4 + 2*x*y - 3*x - 6*y**2 - 5
v1 = sp.integrate(expr1, (y,1,7), (x,-3,4))
v1
v2 = sp.integrate(x**2 * y * sp.exp(x*y), (x,0,1), (y,1,2))
v2
v3 = sp.integrate(x*sp.cos(x**2 + y), (y,0,sp.pi), (x,-sp.sqrt(sp.pi)/2,0))
v3
v3 = sp.integrate(x*sp.cos(x**2 + y), (y,0,sp.pi), (x,-sp.sqrt(sp.pi)/2,0))
v3
```

Zadanie 2 – współrzędne biegunowe

```python
r, phi = sp.symbols('r phi', positive=True)

v5 = sp.integrate((1 - r**2)/(1 + r**2) * r,
                  (r,1,2),
                  (phi, sp.pi/2, 3*sp.pi/2))
v5

v6 = sp.integrate(sp.log(1+r**2)*r,
                  (r,0,1),
                  (phi,-sp.pi/4,3*sp.pi/4))
v6
```

Zadanie 3 – wartość średnia


```python
I = sp.integrate((5-x**2-y**2),
                 (r,0,1),
                 (phi,0,2*sp.pi),
                 deps={x:r*sp.cos(phi), y:r*sp.sin(phi)})

fsr = I/sp.pi
fsr
```

 Zadanie 4 – pole obszaru

```python
pole = sp.integrate((x+1) - 2*x, (x,0,1))
pole
```

Zadanie 5 – objętość bryły

```python
V = sp.integrate(x**2 + y**2, (y,0,4-x), (x,1,4))
V
```

 Zadanie 6 – pole powierzchni

```python
exprS = sp.sqrt(1 + 4 + 9)

S = sp.integrate(exprS, (y,0,(x-5)), (x,2,4))
S
```

 Zadanie 7 – momenty, środek masy

```python
m  = sp.integrate(1, (x,1,5), (y,3,8))
MSx = sp.integrate(y, (x,1,5), (y,3,8))
MSy = sp.integrate(x, (x,1,5), (y,3,8))
Ix = sp.integrate(y**2, (x,1,5), (y,3,8))
Iy = sp.integrate(x**2, (x,1,5), (y,3,8))
IO = sp.integrate(x**2 + y**2, (x,1,5), (y,3,8))

xc = MSy/m
yc = MSx/m

m, MSx, MSy, Ix, Iy, IO, xc, yc
```
